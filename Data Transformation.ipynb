```python
import pandas as pd
import numpy as np
```

```python
Begin_inventory = pd.read_csv("begin_inventory_cleaned.csv")
End_inventory = pd.read_csv("end_inventory_clened.csv")
Purchase_price = pd.read_csv("purchase_price_cleaned.csv")
Purchases = pd.read_csv("purchases_cleaned.csv")
Sales = pd.read_csv("sales_cleaned.csv")
Vendor = pd.read_csv("vendor_cleaned.csv")
```

What is the total sales amount? 

```python
Total_Sales = Sales["Salesprice"].sum()
print(np.around(Total_Sales,2))
```

What is the total Purchase amount?

```python
Total_Purchase = Purchases["Purchaseprice"].sum()
print(np.around(Total_Purchase,2))
```

Which brand has the highest inventory value? 

```python
End_inventory["Inventory_value"] = End_inventory["Onhand"] * End_inventory["Price"]
hst_value = End_inventory.groupby("Description")["Inventory_value"].sum()
print(hst_value.idxmax())
```

Which city has the maximum stock? 

```python
max_stck = End_inventory.groupby("City")["Onhand"].sum()
print(max_stck.idxmax())
print(max_stck.max())
```

Which store has the highest inventory value? 

```python
hst_str =End_inventory.groupby("Store")["Inventory_value"].sum()
print("Store: ",hst_str.idxmax())
print(hst_str.max())
```

Which product has the highest inventory value

```python
hst_prod = End_inventory.groupby("Description")["Inventory_value"].sum()
print(hst_prod.idxmax())
print(hst_prod.max())
```

What is the average inventory value per store? 

```python
avg_store = End_inventory.groupby("Store")["Inventory_value"].mean()
print(np.around(avg_store,2))
```

Which product has the highest price? 

```python
pdt_hst = End_inventory.loc[End_inventory["Price"].idxmax()]
print(pdt_hst["Description"])
print(pdt_hst["Price"])
```

Which size category has the highest inventory quantity? 

```python
hst_invt_quant = End_inventory.groupby("Size")["Onhand"].sum().head(10)
print(hst_invt_quant)
```

13.	Beginning vs Ending Inventory KPIs

```python
Begin_inventory["Inventory_value"] = Begin_inventory["Onhand"] * Begin_inventory["Price"]
End_inventory["Inventory_value"] = End_inventory["Onhand"] * End_inventory["Price"]
```

```python
# Inventory value
Begin_invent = Begin_inventory["Inventory_value"].sum()
End_invent = End_inventory["Inventory_value"].sum()
change_value = End_invent - Begin_invent
print(np.around(change_value,2))
change_pct = (change_value/ Begin_invent) * 100
print(np.around(change_pct,2))
```

```python
# Inventory quantity
Begin_quant = Begin_inventory["Onhand"].sum()
End_quant = End_inventory["Onhand"].sum()
change_quant = End_quant - Begin_quant
print(change_quant)
change_quant_pct = (change_quant / Begin_quant) * 100
print(np.around(change_quant_pct,2))
```

Which city has the highest ending inventory? 

```python
city_end_invt = End_inventory.groupby("City")["Inventory_value"].sum().sort_values(ascending = False).head(5)
print(np.around(city_end_invt,2))
```
