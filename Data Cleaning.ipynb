```python
import pandas as pd
import numpy as np
```

```python
sales = pd.read_csv("Raw/sales.csv")
begin_inventory = pd.read_csv("Raw/begin_inventory.csv")
end_inventory = pd.read_csv("Raw/end_inventory.csv")
purchase_price = pd.read_csv("Raw/purchase_prices.csv")
purchases = pd.read_csv("Raw/purchases.csv")
vendor = pd.read_csv("Raw/vendor_invoice.csv")
```

# Handling Nulls

```python
end_inventory.isnull().sum()
```

```python
end_inventory["City"] = end_inventory["City"].fillna("Unknown")
```

```python
end_inventory.isnull().sum()
```

```python
purchase_price.isnull().sum()
```

```python
purchase_price = purchase_price.dropna(subset=["Description"])
```

```python
purchase_price.isnull().sum()
```

```python
purchases.isnull().sum()
```

```python
vendor["Approval"] = vendor["Approval"].fillna("Pending")
vendor
vendor.isnull().sum()
```

# Remove Extra Spaces and Case type

```python
sales.columns = sales.columns.str.strip().str.title()
begin_inventory.columns = begin_inventory.columns.str.strip().str.title()
end_inventory.columns = end_inventory.columns.str.strip().str.title()
purchase_price.columns = purchase_price.columns.str.strip().str.title()
purchases.columns = purchases.columns.str.strip().str.title()
vendor.columns = vendor.columns.str.strip().str.title()
```

# Saving the cleaned dataset

```python
#sales.to_csv("Sales_clean.csv")
begin_inventory.to_csv("begin_inventory_cleaned.csv",index=False,sep=",",encoding="utf-8")
end_inventory.to_csv("end_inventory_clened.csv",index=False,sep=",",encoding="utf-8")
purchase_price.to_csv("purchase_price_cleaned.csv",index=False,sep=",",encoding="utf-8")
purchases.to_csv("purchases_cleaned.csv",index=False,sep=",",encoding="utf-8")
vendor.to_csv("vendor_cleaned.csv",index=False,sep=",",encoding="utf-8")
sales.to_csv("sales_cleaned.csv",index=False,sep=",",encoding="utf-8")
```
